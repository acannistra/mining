```python
import pandas as pd
import geopandas as gpd
import folium
from shapely.geometry import Point
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline

mines = pd.read_csv("mines_data/rdbms-tab-all/MRDS.txt", sep='\t').set_index("dep_id")
_comm = pd.read_csv("mines_data/rdbms-tab-all/Commodity.txt", sep='\t').set_index('dep_id')
mines = mines.join(_comm.drop('i', axis=1))
```

---
## Economics of Mining
### 1. Land Ownership:

```python
land_status = pd.read_csv("mines_data/rdbms-tab-all/Land_status.txt", sep='\t').set_index('dep_id').drop('i', axis=1)
mines_owners = mines.join(land_status, how='left', rsuffix="__")
mines_owners.head(5)
```

```python
top10groupnames = mines_owners.groupby("commod_group").count()['i'].sort_values(ascending=False)[:10].keys()
top10groups = mines_owners[mines_owners.commod_group.isin(top10groupnames)]
```

```python
sns.set(font_scale=1.2)
sns.factorplot(y="land_st", kind='count', col="commod_group", data=top10groups, col_wrap=5, col_order=top10groupnames)
plt.savefig("landtypetop10.png")
```
